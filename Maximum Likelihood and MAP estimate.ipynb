```python
# Importing header files to use them in our solution
import numpy as np
import scipy.stats as stats
import pandas as pd
```

```python
# Reading the dataset Baseball_data_2005.csv
df = pd.read_csv('Baseball_data_2005.csv',header=None);
# Selecting the last six columns which have successful hits
n_h = df.iloc[1:,10:].values
# We are changing the data type of the data into float
n_hi = n_h.astype('float') 
# We are doing the sum operation on the data which gives us the total number of successful hits
nho=n_hi.sum(axis=1)
# Selecting the column which has the total number of batting attempts
n=df.iloc[1:,3].values
# We are changing the data type of the data into float
ni = n.astype('float')
# We are calculating the maximum likelihood arthimatic way (Number of Hits/Number of Total Trials)
thi=nho/ni
print(thi)
```

```python
# The values of Ho & No are already provided in the question 
Ho=100
No=300
# We are finding the MAP estimate by formula ((h_n+ho)/(ho+no+N))
hoo=Ho+nho
noo=Ho+No+ni
distu=hoo/noo
print(distu)
```
