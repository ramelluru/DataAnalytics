```python
# Importing Libraries Pandas,Sklearn,Numpy,Matplot 
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("url-data.txt")
df1=df.iloc[:,0:]
df1_nm=StandardScaler().fit_transform(df1)
```

```python
## First Approach 
## We directly plugin Varience Percentage in the in build function and finding the Number of Components
pca=PCA(0.95)
principalcomponents=pca.fit_transform(df1_nm)
pca1=pca.explained_variance_ratio_
mi=pca.n_components_
print("Minimum Number of Components for 95%:",mi)

pca=PCA(0.99)
principalcomponents=pca.fit_transform(df1_nm)
pca1=pca.explained_variance_ratio_
mi=pca.n_components_
print("Minimum Number of Components for 99%:",mi)
```

```python
## Second Approach 
## We will plugin the number of components by trial and error method and start finding the desired varience ratio
## Substituting the number of components in the place of n_components
pca=PCA(n_components=11)
principalcomponents=pca.fit_transform(df1_nm)
### how much varience is maintained after projection with 'n ' principal components
pca1=pca.explained_variance_ratio_
## We are adding up to validate the percentage of varience 
pca2=pca1.sum()
print("The Number of principal components:",pca.n_components)
print("The Varience percentage obtained:",pca2)

## Substituting the number of components in the place of n_components
pca=PCA(n_components=22)
principalcomponents=pca.fit_transform(df1_nm)
### how much varience is maintained after projection with 'n ' principal components
pcaf=pca.explained_variance_ratio_
## We are adding up to validate the percentage of varience 
print("Question:1a")
pca2=pcaf.sum()
print("The Number of principal components:",pca.n_components)
print("The Varience percentage obtained:",pca2)
# plotting the varience distribution at each principal component
plt.plot(pcaf)
print("Question:1b")
plt.title('Scree plot for 0.95 Variance with Standardized Data')
plt.xlabel('Number of components')
plt.ylabel('Eigen Values')
plt.show()

#Using the Non Standardized Data
pca=PCA(n_components=13)
principalcomponents=pca.fit_transform(df1)
### how much varience is maintained after projection with 'n ' principal components
pcaf=pca.explained_variance_ratio_
## We are adding up to validate the percentage of varience 
print("Question:1a")
pca2=pcaf.sum()
print("The Number of principal components:",pca.n_components)
print("The Varience percentage obtained:",pca2)
# plotting the varience distribution at each principal component
plt.plot(pcaf)
print("Question:1b")
plt.title('Scree plot for 0.95 Variance with Non-Standardized Data/Raw Data')
plt.xlabel('Number of components')
plt.ylabel('Eigen Values')
plt.show()



pca=PCA(n_components=30)
principalcomponents=pca.fit_transform(df1_nm)
### how much varience is maintained after projection with 'n ' principal components
pcan=pca.explained_variance_ratio_
## We are adding up to validate the percentage of varience
print("Question:1a")
pca2=pcan.sum()
print("The Number of principal components:",pca.n_components)
print("The Varience percentage obtained:",pca2)
# plotting the varience distribution at each principal component
print("Question:1b")
plt.plot(pcan)
plt.title('Scree plot for 0.99 Variance with Standardized Data')
plt.xlabel('Number of components')
plt.ylabel('Eigen Values')
plt.show()

#Using the Non Standardized Data
pca=PCA(n_components=22)
principalcomponents=pca.fit_transform(df1)
### how much varience is maintained after projection with 'n ' principal components
pcan=pca.explained_variance_ratio_
## We are adding up to validate the percentage of varience 
print("Question:1a")
pca2=pcan.sum()
print("The Number of principal components:",pca.n_components)
print("The Varience percentage obtained:",pca2)
# plotting the varience distribution at each principal component
print("Question:1b")
plt.plot(pcan)
plt.title('Scree plot for 0.99 Variance with Non Standardized Data/Raw Data')
plt.xlabel('Number of components')
plt.ylabel('Eigen Values')
plt.show()
```

```python
## Third Approach 
## We will plugin the number of components by trial and error method in the loop with standardized data and 
##start printing the desired varience ratio which achieve minimum number of components
pca2=[]
for i in range(60): 
    pca=PCA(n_components=i) 
    principalcomponents=pca.fit_transform(df1_nm)
    pca1=pca.explained_variance_ratio_
    spca=round(pca.explained_variance_ratio_.sum(),2)
    #Prints the Minimum number of components when varience is 95%
    if spca==0.95:
        print("Varience is",spca)
        print("Minimum number of components is:",i)
    #Prints the Minimum number of components when varience is in between 95% and 99%
    if spca>0.95 and spca<0.99:
        print("Varience is",spca)
        print("Minimum number of components is:",i)
    #Prints the Minimum number of components when varience is 99%
    if spca==0.99:
        print("Varience is",spca)
        print("Minimum number of components is:",i)
    pca2.append(spca)
```
